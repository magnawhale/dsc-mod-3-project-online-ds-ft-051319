## Importing Data & Libraries

```python
# For SQL & dataframes
import pandas as pd
import sqlite3

# For math & statistics
import numpy as np
import scipy.stats as stats

# For graphing
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid")
```

```python
conn = sqlite3.connect('Northwind_small.sqlite')
c = conn.cursor()

# Let's get a list of all of the tables present in our database
table_names = c.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
table_names
```

## Make some dataframes

### Employees

```python
c.execute("select * from 'Employee';")
employee_df = pd.DataFrame(c.fetchall())
employee_df.columns = [x[0] for x in c.description]
display(employee_df)
print('employee_df\n',employee_df.info())

c.execute("select * from 'EmployeeTerritory';")
employeeterritory_df = pd.DataFrame(c.fetchall())
employeeterritory_df.columns = [x[0] for x in c.description]
display(employeeterritory_df.head(10))
print('employeeterritory_df\n',employeeterritory_df.info())

c.execute("select * from 'Territory';")
territory_df = pd.DataFrame(c.fetchall())
territory_df.columns = [x[0] for x in c.description]
display(territory_df.head(10))
print('territory_df\n',territory_df.info())

c.execute("select * from 'Region';")
region_df = pd.DataFrame(c.fetchall())
region_df.columns = [x[0] for x in c.description]
display(region_df.head(10))
print('region_df\n',region_df.info())
```

```python
# Check for NaNs
print(employee_df.isna().sum())
print(employeeterritory_df.isna().sum())
print(territory_df.isna().sum())
print(region_df.isna().sum())
```

```python
# Check for placeholder values and other weird enries

for col in employee_df.columns:
    print(col, '\n', employee_df[col].value_counts(normalize=True).head())
    print(col, '\n', employee_df[col].value_counts(normalize=True).tail(), '\n\n')

for col in employeeterritory_df.columns:
    print(col, '\n', employeeterritory_df[col].value_counts(normalize=True).head())
    print(col, '\n', employeeterritory_df[col].value_counts(normalize=True).tail(), '\n\n')
    
for col in territory_df.columns:
    print(col, '\n', territory_df[col].value_counts(normalize=True).head())
    print(col, '\n', territory_df[col].value_counts(normalize=True).tail(), '\n\n')

for col in region_df.columns:
    print(col, '\n', region_df[col].value_counts(normalize=True).head())
    print(col, '\n', region_df[col].value_counts(normalize=True).tail(), '\n\n')
```

#### Analysis of Employee data

These tables seem to be in pretty good shape. The only relevant NaNs for employees are in the reporting structure for the VP, who does not report to anyone. However, one glaring error is that all of the `HireDate` values are at least five years in the future. Consequently, we will not be able to use any information from this column.

Looking a bit deeper, it seems that we have nine employees at different levels within the company hierarchy. If we ask any employee-related questions, we may need to keep this information in mind. Additionally, the employees live in two different countries, which may also have an effect on any questions we ask.
